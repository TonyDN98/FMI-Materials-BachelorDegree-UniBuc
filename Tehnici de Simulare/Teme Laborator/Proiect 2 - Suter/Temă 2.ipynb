```python
### Importăm librăriile necesare pe tot parcursul temei
import math      # pentru e
import random    
import numpy as np
```

```python
"""
    Rezolvare 1: 
        1) Conform cursului 6 (slide 3), dacă Y ~ Gama(0, 1, a) și Z ~ Gama(0, 1, b) sunt 2 variabile 
    independente, atunci variabila X = Y / (Y + Z) este o variabilă Beta(a, b). Librăria 'random' 
    din Python vine cu funcția 'gammavariate', care este folosită pentru a returna un număr aleatoriu, 
    în virgulă mobilă, cu distribuție gamma.
        2) Conform cursului 6 (slide 7, nu 5, deoarece ni se cere Beta(0.3, 5), iar 5 > 1, nu 5 < 1), 
    având 0 < a < 1, b > 1 și U1, U2 2 variabile aleatoare uniforme, independente pe mulțimea [0, 1], 
    dacă Y = U1^(1/a) și Z = U2^(1/(b-1)), atunci repartiția variabilei Y condiționată de Y + Z < 1 este 
    Beta(a, b). Algoritmul care trebuie implementat este foarte similar cu cel din cursul 6, slide 6. 
    Librăria 'random' din Python vine cu funcția 'uniform', care generează numere dintr-o distribuție uniformă. 
    
    
    Bibliografie:
    Cursul 6
    https://interactivechaos.com/en/python/function/randomgammavariate                 -> Pentru random.gammavariate
    https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html  -> Pentru random.uniform
"""





# Metoda 1 de generare a Variabilei Beta
def Beta1(a, b):
    Y = random.gammavariate(a, 1)     # conform documentației, a și 1 trebuie să fie ambele "> 0"
    Z = random.gammavariate(b, 1)
    
    X = Y / (Y + Z)                   # variabila Beta cerută
    return X



# Metoda 2 de generare a Variabilei Beta -> Folosim această metodă deoarece generarea a 2 variabile Gama crește 
# considerabil complexitatea programului și cerința noastră se pliază pe unul dintre cazurile particulare
def Beta2(a, b):
    while True:                    # repetăm până condiția se adeverește
        U1 = random.uniform(0, 1)
        U2 = random.uniform(0, 1)
    
        Y = U1 ** (1 / a)
        Z = U2 ** (1 / (b - 1))
    
        if Y + Z < 1:
            return Y               # variabila Beta cerută
```

```python
"""
    Conform cerinței, vom realiza o verificare folosind media și dispersia de selecție și vom compara valorile teoretice cu
    cele generate de noi. Pentru acest lucru, vom genera un număr mare de valori.
    
    Beta(a, b)
    Media Teoretică: a / (a + b)
    Deviația Standard Teoretică: sqrt((a * b) / (((a + b)**2) * (a + b + 1)))       ( = sqrt(Variație))
    
    Pentru numerele generate de noi, folosim funcțiile 'mean' (media) și 'std' (deviația standard) din biblioteca 'numpy'.
    
    Bibliografie: 
    Curs 6, slide 2
    https://en.wikipedia.org/wiki/Beta_distribution
    https://numpy.org/doc/stable/reference/generated/numpy.mean.html    -> Media
    https://numpy.org/doc/stable/reference/generated/numpy.std.html     -> Deviația Standard
"""


### Date Generale
a = 0.3
b = 5
nr_simulari = 100000


### Formulele Teoretice pt Medie și Deviația Standard
media_teoretica = a / (a + b)
deviatiaStandard_teoretica = np.sqrt((a * b) / (((a + b)**2) * (a + b + 1)))


### Generăm valori și calculăm media și deviația pe baza acestora
set_Beta1 = [Beta1(a, b) for i in range(nr_simulari)]
set_Beta2 = [Beta2(a, b) for i in range(nr_simulari)]

media_generata_Beta1 = np.mean(set_Beta1)
media_generata_Beta2 = np.mean(set_Beta2)

deviatiaStandard_Beta1 = np.std(set_Beta1)
deviatiaStandard_Beta2 = np.std(set_Beta2)


### Comparăm
print(f'Media Teoretică a Variabilei Beta(0.3, 5): {media_teoretica}')
print(f'Media Generată a Variabilei Beta(0.3, 5) prin Metoda 1: {media_generata_Beta1}')
print(f'Media Generată a Variabilei Beta(0.3, 5) prin Metoda 2: {media_generata_Beta2}\n')

print(f'Deviația Standard Teoretică a Variabilei Beta(0.3, 5): {deviatiaStandard_teoretica}')
print(f'Deviația Standard Generată a Variabilei Beta(0.3, 5) prin Metoda 1: {deviatiaStandard_Beta1}')
print(f'Deviația Standard Generată a Variabilei Beta(0.3, 5) prin Metoda 2: {deviatiaStandard_Beta2}')
```

```python
"""
    Rezolvare 2: 
        Conform cursului 6 (slide 14), dacă variabilele U1, U2 sunt uniforme pe [0, 1] și independente, atunci variabilele
    aleatoare Z1 = V1 * sqrt((-2 * log S) / S), Z2 = V2 * sqrt((-2 * log S) / S), cu V1 = 2U1 - 1, V2 = 2U2 - 1, 
    S = V1^2 + V2^2, S < 1, sunt variabile N(0, 1) independente.
    
    
    Bibliografie:
    Cursul 6
"""



def MetodaPolara():
    while True:                       # repetăm până condiția se adeverește
        U1 = random.uniform(0, 1)
        U2 = random.uniform(0, 1)
    
        V1 = 2 * U1 - 1
        V2 = 2 * U2 - 1
        S = V1 ** 2 + V2 ** 2
    
        if S < 1:                                      # se resping valorile pt care S >= 1, iar prob de acceptare = pi/4
            # Z1 și Z2 pt fi simulate și cu alte formule (curs 6, slide 19), dar, deși nu mai apar respingeri, complexitatea
            # crește datorită faptului că folosim funcții trigonometrice și logaritmi, deci folosim metoda prezentată mai jos
            Z1 = V1 * np.sqrt((-2 * np.log(S)) / S)    # se obțin simultan 2 valori de selecție variabile N(0, 1) independente
            Z2 = V2 * np.sqrt((-2 * np.log(S)) / S)
            
            return Z1, Z2
```

```python
"""
    Conform cerinței, vom realiza o verificare folosind media și dispersia de selecție și vom compara valorile teoretice cu
    cele generate de noi. Pentru acest lucru, vom genera un număr mare de valori.
    
    N(a, b)                                    -> în cazul nostru, N(0, 1)
    Media Teoretică: a                         -> 0
    Deviația Standard Teoretică: sqrt(b)       -> 1
    
    Pentru numerele generate de noi, folosim funcțiile 'mean' (media) și 'std' (deviația standard) din biblioteca 'numpy'.
    
    Bibliografie: 
    https://en.wikipedia.org/wiki/Normal_distribution
    https://numpy.org/doc/stable/reference/generated/numpy.mean.html  -> Media
    https://numpy.org/doc/stable/reference/generated/numpy.std.html   -> Deviația Standard
"""


### Date Generale
a = 0
b = 1
nr_simulari = 100000


### Formulele Teoretice pt Medie și Deviația Standard
media_teoretica = a
deviatiaStandard_teoretica = np.sqrt(b)


### Generăm valori și calculăm media și deviația pe baza acestora
set_N = []
for i in range(nr_simulari):
    Z1, Z2 = MetodaPolara()
    set_N.append(Z1)
    set_N.append(Z2)

media_generata = np.mean(set_N)
deviatiaStandard_generata = np.std(set_N)


### Comparăm
print(f'Media Teoretică: {media_teoretica}')
print(f'Media Generată prin Metoda Polară: {media_generata}\n')

print(f'Deviația Standard Teoretică: {deviatiaStandard_teoretica}')
print(f'Deviația Standard Generată prin Metoda Polară: {deviatiaStandard_generata}')
```

```python
"""
    Rezolvare 1: 
        Algoritmul se găsește în cursul 5, slide-ul 6. Formulele pentru medie și dispersie se găsesc în cursul 5, slide-ul 2.
    
    
    Bibliografie:
    Cursul 5
    https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html -> Pentru random.uniform
"""



# Generare Variabilă Exponențială folosind a treia teoremă de respingere
def Respingere3_Exp():
    N = 0                               # nr aleator de subșiruri descendente respinse până se îndeplinește condiția
    while True:                         # repetăm până condiția se adeverește
        U0 = random.uniform(0, 1)
        U1 = random.uniform(0, 1)
        U_star = U0
        K = 1                          # K = indicele aleator la care se "rupe" șirul descendent
    
        while U0 >= U1:
            K += 1
            U0 = U1
            U1 = random.uniform(0, 1)
    
        if K%2 == 1:
            X = N + U_star
            return X

        N += 1
```

```python
"""
    Rezolvare 2: 
        1) Primul agoritm se află în cursul 7, slide-ul 19.
        2) Al doilea agoritm se află în cursul 7, slide-ul 20.
    
    Repartiția Poisson(a) este folosită pentru a modela nr de evenimente care apar într-un interval de timp. 'a' = nr mediu de 
    evenimente dintr-un anumit interval d timp.
    
    Bibliografie:
    Cursul 7
    https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html   -> Pentru random.uniform
    https://numpy.org/doc/stable/reference/random/generated/numpy.random.binomial.html  -> Pentru random.binomial
"""





# Metoda 1 de Generare a Variabilei Poisson
def Poisson1(a):
    i = 0
    P = 1
    
    while True:                      # repetăm până condiția se adeverește
        U = random.uniform(0, 1)
        i += 1
        P *= U
        
        if P < math.exp(-a):
            X = i - 1
            return X



# Metoda 2 de Generare a Variabilei Poisson
def Poisson2(a):                    # a = np
    p = 0.001                       # p -> 0 (tinde la)
    n = int(round(a / p))           # n -> inf (tinde la)
    X = np.random.binomial(n, p)
    return X
```

```python
"""
    Conform cerinței, vom realiza o verificare folosind media și dispersia de selecție și vom compara valorile teoretice cu
    cele generate de noi. Pentru acest lucru, vom genera un număr mare de valori.
   
    Poisson(1)                                    
    Media Teoretică: 1
    Deviația Standard Teoretică: 1
    
    Pentru numerele generate de noi, folosim funcțiile 'mean' (media) și 'std' (deviația standard) din biblioteca 'numpy'.
    
    Bibliografie: 
    Curs 7
    https://en.wikipedia.org/wiki/Poisson_distribution
    https://numpy.org/doc/stable/reference/generated/numpy.mean.html  -> Media
    https://numpy.org/doc/stable/reference/generated/numpy.std.html   -> Deviația Standard
"""


### Date Generale
nr_simulari = 100000


### Formulele Teoretice pt Medie și Deviația Standard
media_teoretica = 1
deviatiaStandard_teoretica = 1


### Generăm valori și calculăm media și deviația pe baza acestora
set_Poisson1 = [Poisson1(1) for i in range(nr_simulari)]
set_Poisson2 = [Poisson2(1) for i in range(nr_simulari)]

media_generata_Poisson1 = np.mean(set_Poisson1)
media_generata_Poisson2 = np.mean(set_Poisson2)

deviatiaStandard_Poisson1 = np.std(set_Poisson1)
deviatiaStandard_Poisson2 = np.std(set_Poisson2)


### Comparăm
print(f'Media Teoretică a Variabilei Poisson(1): {media_teoretica}')
print(f'Media Generată a Variabilei Poisson(1) prin Metoda 1: {media_generata_Poisson1}')
print(f'Media Generată a Variabilei Poisson(1) prin Metoda 2: {media_generata_Poisson2}\n')

print(f'Deviația Standard Teoretică a Variabilei Poisson(1): {deviatiaStandard_teoretica}')
print(f'Deviația Standard Generată a Variabilei Poisson(1) prin Metoda 1: {deviatiaStandard_Poisson1}')
print(f'Deviația Standard Generată a Variabilei Poisson(1) prin Metoda 2: {deviatiaStandard_Poisson2}')
```
